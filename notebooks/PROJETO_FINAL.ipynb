- ponto de origem `A`;
- ponto de destino `B`;
- distância máxima de caminhada `X`;

o grupo deve escolher um ponto de embarque `P`, tal que `P` esteja a no máximo `X` metros de caminhada da origem `A`.

A rota completa será composta por dois trechos:

```text
A → P  caminhada
P → B  carro
```

#Implementação dos Algoritmos

````python
def heuristica(node_a, node_b, G):
    """Calcula a distância em linha reta (Great Circle) entre dois nós."""
    return ox.distance.great_circle(
        G.nodes[node_a]["y"], G.nodes[node_a]["x"],
        G.nodes[node_b]["y"], G.nodes[node_b]["x"]
    )

def dijkstra_simples_projeto(G, origem, destino, weight="length"):
    inicio = time.time()
    dist = {node: float("inf") for node in G.nodes()}
    predecessores = {node: None for node in G.nodes()}
    dist[origem] = 0
    visitados = set()
    nos_expandidos = 0

    while len(visitados) < len(G):
        atual = None
        menor = float("inf")

        for node in G.nodes():
            if node not in visitados and dist[node] < menor:
                menor = dist[node]
                atual = node

        if atual is None or atual == destino:
            if atual == destino:
                nos_expandidos += 1
            break

        visitados.add(atual)
        nos_expandidos += 1

        for vizinho in G.neighbors(atual):
            if G.is_multigraph():
                peso = min(data.get(weight, 1) for data in G[atual][vizinho].values())
            else:
                peso = G[atual][vizinho].get(weight, 1)

            nova_dist = dist[atual] + peso
            if nova_dist < dist[vizinho]:
                dist[vizinho] = nova_dist
                predecessores[vizinho] = atual

    caminho = []
    atual = destino
    if predecessores[atual] is not None or atual == origem:
        while atual is not None:
            caminho.insert(0, atual)
            atual = predecessores[atual]
    return caminho, dist[destino], nos_expandidos, (time.time() - inicio)

def dijkstra_heap_projeto(G, origem, destino, weight="length"):
    inicio = time.time()
    dist = {node: float("inf") for node in G.nodes()}
    predecessores = {node: None for node in G.nodes()}
    dist[origem] = 0
    fila = [(0, origem)]
    nos_expandidos = 0

    while fila:
        distancia_atual, atual = heapq.heappop(fila)
        nos_expandidos += 1

        if atual == destino:
            break

        if distancia_atual > dist[atual]:
            continue

        for vizinho in G.neighbors(atual):
            if G.is_multigraph():
                peso = min(data.get(weight, 1) for data in G[atual][vizinho].values())
            else:
                peso = G[atual][vizinho].get(weight, 1)

            nova = distancia_atual + peso
            if nova < dist[vizinho]:
                dist[vizinho] = nova
                predecessores[vizinho] = atual
                heapq.heappush(fila, (nova, vizinho))

    caminho = []
    atual = destino
    if predecessores[atual] is not None or atual == origem:
        while atual is not None:
            caminho.insert(0, atual)
            atual = predecessores[atual]

    return caminho, dist[destino], nos_expandidos, (time.time() - inicio)

def astar_customizado_projeto(G, origem, destino, weight="length"):
    inicio = time.time()
    dist = {node: float("inf") for node in G.nodes()}
    predecessores = {node: None for node in G.nodes()}
    dist[origem] = 0

    fila = [(0 + heuristica(origem, destino, G), 0, origem)]
    nos_expandidos = 0

    while fila:
        _, distancia_atual, atual = heapq.heappop(fila)
        nos_expandidos += 1

        if atual == destino:
            break

        if distancia_atual > dist[atual]:
            continue

        for vizinho in G.neighbors(atual):
            if G.is_multigraph():
                peso = min(data.get(weight, 1) for data in G[atual][vizinho].values())
            else:
                peso = G[atual][vizinho].get(weight, 1)

            nova_dist = distancia_atual + peso

            if nova_dist < dist[vizinho]:
                dist[vizinho] = nova_dist
                predecessores[vizinho] = atual

                prioridade = nova_dist + heuristica(vizinho, destino, G)
                heapq.heappush(fila, (prioridade, nova_dist, vizinho))

    caminho = []
    atual = destino
    if predecessores[atual] is not None or atual == origem:
        while atual is not None:
            caminho.insert(0, atual)
            atual = predecessores[atual]

    return caminho, dist[destino], nos_expandidos, (time.time() - inicio)

def bidirecional_manual_projeto(G, origem, destino, weight="length"):
    inicio = time.time()
    dist_f = {node: float("inf") for node in G.nodes()}
    dist_b = {node: float("inf") for node in G.nodes()}
    pai_f = {node: None for node in G.nodes()}
    pai_b = {node: None for node in G.nodes()}

    dist_f[origem] = 0
    dist_b[destino] = 0
    fila_f = [(0, origem)]
    fila_b = [(0, destino)]
    visitados_f = set()
    visitados_b = set()

    melhor_custo = float("inf")
    no_encontro = None
    nos_expandidos = 0

    while fila_f and fila_b:
        if fila_f[0][0] + fila_b[0][0] >= melhor_custo:
            break

        d_f, u = heapq.heappop(fila_f)
        nos_expandidos += 1
        visitados_f.add(u)
        for vizinho in G.neighbors(u):
            if G.is_multigraph():
                peso = min(data.get(weight, 1) for data in G[u][vizinho].values())
            else:
                peso = G[u][vizinho].get(weight, 1)

            nova = dist_f[u] + peso
            if nova < dist_f[vizinho]:
                dist_f[vizinho] = nova
                pai_f[vizinho] = u
                heapq.heappush(fila_f, (nova, vizinho))
            if vizinho in visitados_b:
                if dist_f[vizinho] + dist_b[vizinho] < melhor_custo:
                    melhor_custo = dist_f[vizinho] + dist_b[vizinho]
                    no_encontro = vizinho

        d_b, v = heapq.heappop(fila_b)
        nos_expandidos += 1
        visitados_b.add(v)
        for vizinho in G.neighbors(v):
            if G.is_multigraph():
                peso = min(data.get(weight, 1) for data in G[v][vizinho].values())
            else:
                peso = G[v][vizinho].get(weight, 1)

            nova = dist_b[v] + peso
            if nova < dist_b[vizinho]:
                dist_b[vizinho] = nova
                pai_b[vizinho] = v
                heapq.heappush(fila_b, (nova, vizinho))
            if vizinho in visitados_f:
                if dist_f[vizinho] + dist_b[vizinho] < melhor_custo:
                    melhor_custo = dist_f[vizinho] + dist_b[vizinho]
                    no_encontro = vizinho

    caminho = []
    if no_encontro is not None:
        atual = no_encontro
        caminho_f = []
        while atual is not None:
            caminho_f.append(atual)
            atual = pai_f[atual]
        caminho_f.reverse()
        atual = pai_b[no_encontro]
        caminho_b = []
        while atual is not None:
            caminho_b.append(atual)
            atual = pai_b[atual]
        caminho = caminho_f + caminho_b

    return caminho, melhor_custo, nos_expandidos, (time.time() - inicio)
````

#Modelagem do Trânsito Sintético

````python
for u, v, k, data in G.edges(keys=True, data=True):
    fator_transito = 1.0

    tipo_via = data.get("highway")

    if tipo_via == "primary":
        fator_transito = 1.2

    elif tipo_via == "secondary":
        fator_transito = 1.5

    else:
        fator_transito = 10

    data["travel_time_congestion"] = data["travel_time"] * fator_transito
````

#Lógica do Ponto de Embarque (RideSmart)

````python
def dijkstra_cutoff_manual(G, source, cutoff, weight="length"):
    """
    Implementação manual do Dijkstra para encontrar distâncias de uma origem
    para todos os nós alcançáveis, respeitando um limite máximo de distância (cutoff).
    """
    dist = {source: 0}

    fila = [(0, source)]

    while fila:
        distancia_atual, atual = heapq.heappop(fila)

        if distancia_atual > cutoff:
            break

        if distancia_atual > dist.get(atual, float('inf')):
            continue

        for vizinho in G.neighbors(atual):
            peso_aresta = G[atual][vizinho][0].get(weight, 1)
            nova_distancia = distancia_atual + peso_aresta

            if nova_distancia <= cutoff:
                if nova_distancia < dist.get(vizinho, float('inf')):
                    dist[vizinho] = nova_distancia
                    heapq.heappush(fila, (nova_distancia, vizinho))

    return dist
````

````python
def rodar_simulacao_ridesmart(G, origem_coords, destino_coords, raio_caminhada=300):
    """
    Executa a otimização RideSmart.
    Garante que as colunas do DataFrame existam e trata cenários vazios.
    """
    origem_node = ox.distance.nearest_nodes(G, origem_coords[1], origem_coords[0])
    destino_node = ox.distance.nearest_nodes(G, destino_coords[1], destino_coords[0])

    candidatos = dijkstra_cutoff_manual(G, origem_node, cutoff=raio_caminhada, weight="length")

    resultados = []
    VELOCIDADE_CAMINHADA = 2

    for ponto_P, dist_caminhada_m in candidatos.items():
        try:
            tempo_caminhada_seg = dist_caminhada_m / VELOCIDADE_CAMINHADA

            _, tempo_carro_seg, _, _ = dijkstra_heap_projeto(
                G, ponto_P, destino_node, weight="travel_time_congestion"
            )

            if tempo_carro_seg == float("inf"): continue

            tempo_total_min = (tempo_caminhada_seg + tempo_carro_seg) / 60

            resultados.append({
                "ponto": ponto_P,
                "distancia_caminhada_metros": dist_caminhada_m,
                "tempo_total_minutos": tempo_total_min
            })
        except:
            pass

    if not resultados:
        return pd.DataFrame(columns=["ponto", "distancia_caminhada_metros", "tempo_total_minutos"]), origem_node, destino_node

    df = pd.DataFrame(resultados).sort_values("tempo_total_minutos").reset_index(drop=True)
    return df, origem_node, destino_node
````

````python
cenarios = {
    "Residencial": ((-5.5820, -36.9150), (-5.5680, -36.8900), 300),
    "Centro": ((-5.5710, -36.9000), (-5.5700, -36.8950), 600)
}

for nome, (origem, destino, distância) in cenarios.items():
    print(f"\n>>> Testando cenário: {nome}")
    df_res, _, _ = rodar_simulacao_ridesmart(G, origem, destino, distância)

    if df_res.empty:
        print("Aviso: Nenhum ponto de embarque válido foi encontrado neste cenário!")
        print("Dica: Tente aumentar o RAIO_MAX_CAMINHADA ou verifique as coordenadas.")
        continue

    melhor = df_res.iloc[0]
    print(f"Melhor ponto de embarque encontrado: {int(melhor['ponto'])}")
    print(f"Tempo total: {melhor['tempo_total_minutos']:.2f} minutos")
    print(f"Distância a pé: {melhor['distancia_caminhada_metros']:.2f} metros")
````

##Comparação dos Algoritmos

````python
cenarios = {
    "Residencial": ((-5.5820, -36.9150), (-5.5680, -36.8900)),
    "Centro": ((-5.5710, -36.9000), (-5.5700, -36.8950))
}

algoritmos = {
    "Dijkstra Simples": dijkstra_simples_projeto,
    "Dijkstra Heap": dijkstra_heap_projeto,
    "A* (Customizado)": astar_customizado_projeto,
    "Bidirecional": bidirecional_manual_projeto
}

resultados_benchmark = []

for cenario, (origem_coords, destino_coords) in cenarios.items():
    o_node = ox.distance.nearest_nodes(G, origem_coords[1], origem_coords[0])
    d_node = ox.distance.nearest_nodes(G, destino_coords[1], destino_coords[0])

    for nome_alg, func in algoritmos.items():
        _, custo_segundos, nos, tempo = func(G, o_node, d_node, weight="travel_time_congestion")

        resultados_benchmark.append({
            "Cenário": cenario,
            "Algoritmo": nome_alg,
            "Tempo de Execução (s)": tempo,
            "Nós Expandidos": nos,
            "Tempo de Viagem (min)": custo_segundos / 60
        })

df_benchmark_final = pd.DataFrame(resultados_benchmark)
````

````python
# Lista para armazenar os dados de todos os cenários para o relatório
dados_experimento = []

# Cenários para testar (Residencial e Centro)
cenarios = {
    "Residencial": ((-5.5820, -36.9150), (-5.5680, -36.8900)),
    "Centro": ((-5.5710, -36.9000), (-5.5700, -36.8950))
}

print(f"{'Cenário':<15} | {'Tempo RideSmart (min)':<22} | {'Tempo Sem RideSmart (min)':<25} | {'Ganho (%)'}")
print("-" * 85)

for nome, (origem, destino) in cenarios.items():
    # 1. Execução RideSmart
    df_res, o_node, d_node = rodar_simulacao_ridesmart(G, origem, destino, raio_caminhada=2000)

    if df_res.empty:
        print(f"{nome:<15} | Erro: Sem pontos válidos")
        continue

    melhor = df_res.iloc[0]
    tempo_ridesmart = melhor['tempo_total_minutos']

    # 2. Execução "Sem Caminhada" (Direto da origem)
    _, tempo_sem_caminhada_seg, _, _ = dijkstra_heap_projeto(
        G, o_node, d_node, weight="travel_time_congestion"
    )
    tempo_sem_caminhada_min = tempo_sem_caminhada_seg / 60

    # 3. Cálculo de Ganho
    ganho = ((tempo_sem_caminhada_min - tempo_ridesmart) / tempo_sem_caminhada_min) * 100

    # Armazenando para gerar gráficos depois
    dados_experimento.append({
        "Cenário": nome,
        "RideSmart": tempo_ridesmart,
        "Sem_RideSmart": tempo_sem_caminhada_min,
        "Ganho_%": ganho
    })

    print(f"{nome:<15} | {tempo_ridesmart:<22.2f} | {tempo_sem_caminhada_min:<25.2f} | {ganho:.2f}%")

# Criar DataFrame final de resultados
df_final = pd.DataFrame(dados_experimento)
````
